```python
data1.isnull().sum()
```

```python
missing_percent = (data1.isnull().sum() / len(data1))*100
missing_percent
```

```python
for i in data1.columns:
    if missing_percent[i]>90:
        print(i)
```

```python
for i in data1.columns:
    if missing_percent[i]>90:
        data1.drop(i, axis=1, inplace=True)
```

```python
data1.columns
```

```python
data1.drop('county_name',axis=1,inplace=True)
```

```python
data1.drop('search_type',axis=1,inplace=True)
```

```python
# drop several columns together

data1.drop(['county_name','search_type'], axis=1, inplace=True)
```

```python
data1[['driver_gender', 'driver_race']] = data1[['driver_gender', 'driver_race']].fillna('Unknown')
```

```python
data1['driver_age'].fillna(value=data1['driver_age'].mean(), inplace=True)
```
